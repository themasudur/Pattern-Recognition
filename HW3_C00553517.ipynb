```python
# 1
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

X1, y1 = make_moons(n_samples=250, noise=0.20, random_state=51)
X2, y2 = make_moons(n_samples=250, noise=0.35, random_state=51)

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.scatter(X1[:, 0], X1[:, 1], c=y1)
plt.title('data (noise=0.20)')
plt.xlim([-2.5, 2.5])
plt.ylim([-2.5, 2.5])

plt.subplot(1, 2, 2)
plt.scatter(X2[:, 0], X2[:, 1], c=y2)
plt.title('data (noise=0.35)')
plt.xlim([-2.5, 2.5])
plt.ylim([-2.5, 2.5])
plt.tight_layout()
plt.show()
```

```python
# 2
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.3, random_state=51)

# Create a knn classifier k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Train the classifier knn.fit(X_train, y_train)
knn.fit(X_train, y_train)

# Make predictions on the test set y_pred = knn.predict(X_test)
y_pred = knn.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"KNN (k=3) Accuracy: {accuracy:.2f}")
```

```python
for k in [1, 2, 4]:
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train, y_train)
  y_pred = knn.predict(X_test)
  accuracy = accuracy_score(y_test, y_pred)
  print(f"KNN (k={k}) Accuracy: {accuracy:.2f}")
```

```python
# 3
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size=0.3, random_state=51)

# Create a knn classifier k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Train the classifier knn.fit(X_train, y_train)
knn.fit(X_train, y_train)

# Make predictions on the test set y_pred = knn.predict(X_test)
y_pred = knn.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"KNN (k=3) Accuracy: {accuracy:.2f}")
```

```python
for k in [1, 2, 4]:
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train, y_train)
  y_pred = knn.predict(X_test)
  accuracy = accuracy_score(y_test, y_pred)
  print(f"KNN (k={k}) Accuracy: {accuracy:.2f}")
```
